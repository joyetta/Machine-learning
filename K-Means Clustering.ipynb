Rescaling 

Data should be between -1 and 1

```python
df_scaled = df[["Age","Annual Income (k$)","Spending Score (1-100)"]]
```

```python
# Class instance
scaler = StandardScaler()
```

```python
# Fit_transform
df_scaled_fit = scaler.fit_transform(df_scaled)
```

```python
df_scaled_fit = pd.DataFrame(df_scaled_fit)
df_scaled_fit.columns = ["Age","Annual Income (k$)","Spending Score (1-100)"]
df_scaled_fit.head()
```

```python
var_list = df_scaled_fit[["Annual Income (k$)","Spending Score (1-100)"]]
```

K-MEANS CLUSTERING MODEL

```python
# Let K be a random number, 4

kmeans = KMeans(n_clusters=4, max_iter=50)
kmeans.fit(var_list)
```

```python
kmeans.labels_
```

```python
# Calculating sum of squared distance 
# the distance between each data point and the line of best fit is squared and then summed up

ssd = []

for num_clusters in range(1,11):
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50)
    kmeans.fit(var_list)
    
    ssd.append(kmeans.inertia_)
```

Elbow curve to discover the optimal number of clusters for K-means model

```python
plt.figure(figsize=(12,6))

plt.plot(range(1,11), ssd, linewidth=2, color="red", marker ="8")
plt.title("Elbow Curve", fontsize=20, color="green")
plt.xlabel("K Value")
plt.xticks(np.arange(1,11,1))
plt.ylabel("SSD")

plt.show()
```

Optimum K is 5, from curve above

```python
kmeans = KMeans(n_clusters=5, max_iter=50)
kmeans.fit(var_list)
```

```python
kmeans.labels_
```

```python
df["Label"] = kmeans.labels_
```

```python
df.head()
```

```python
# scatter plot when k = 5 (optimal k)

plt.figure(figsize=(10,6))

plt.title("Ploting the data into 5 clusters", fontsize=20, color="green")
sns.scatterplot(data=df, x="Annual Income (k$)", 
                y="Spending Score (1-100)", hue="Label", s=60, palette=['green','orange','brown','blue','red'])


plt.show()
```

```python
# showing the clusters

plt.figure(figsize=(10,6))
sns.boxplot(x='Label', y='Annual Income (k$)', data=df, palette="nipy_spectral_r")
plt.title("Label Wise Customer's Income", fontsize=20, color="green")
plt.xlabel(xlabel="Label", fontsize=15)
plt.ylabel(ylabel="Annual Income (k$)",fontsize=15)
plt.show()
```
